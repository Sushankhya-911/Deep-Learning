```python
#Load a BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

#Example input sentences
sentence1 = "The cat sat on the mat."
sentence2 = "The dog barked loudly."

#Tokenize the input sentences
tokenized_input = tokenizer(sentence1, sentence2, return_tensors="pt", padding=True, truncation=True)

#Access token type IDs
token_type_ids = tokenized_input["token_type_ids"]

print(token_type_ids)
```

#Compile the model

This model is only used for tenorflow

```python
from transformers import TFAutoModelForSequenceClassification
from tensorflow.keras.optimizers import Adam
#Load and compile model
model=TFAutoModelForSequenceClassification.from_pretrained('google-bert/bert-base-cased')
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy')
model.summary()
```

#Fit the model

```python
#x=tokenized_data
#y=labels
import numpy as np
labels = np.array(dataset["label"])
model.fit(tokenized_data,labels)
```
